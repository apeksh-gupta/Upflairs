1. Write a program to find the minimum price of ktm bike, 
which are available in Jaipur, and how many ktm bikes are 
available at Jaipur location

```python
import pandas as pd
df = pd.read_csv('Used_Bikes.csv')
a = df[ (df['brand'] == "KTM")  &  (df['city'] == "Jaipur") ]
a
```

```python
a.shape  ## 8 ktm  bikes available in jaipur 
```

```python
a['price'].min()  ## minimum price 
```

2. Find out how many bikes are available at Delhi location, and 
that are “first owner” bikes. And also after filter the data 
find out the minimum and maximum age of the bike.

```python
import pandas as pd
df = pd.read_csv('Used_Bikes.csv')
a = df[ (df['owner'] == "First Owner")  &  (df['city'] == "Delhi") ]
a    ## 6516 bikes are available in Delhi which are first hand Owner
```

```python
a['age'].min()      # Minumum age of a bike is 1 
```

```python
a["age"].max()      # Maximum age of a bike is 20
```
